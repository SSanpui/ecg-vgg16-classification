```python
!pip install pydot
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dropout, Flatten, Dense, BatchNormalization, GlobalAveragePooling2D
from tensorflow.keras.utils import plot_model

# Define the new CNN model with improvements
model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(222, 222, 3), name='conv2d_1'),
    BatchNormalization(name='batch_norm_1'),
    MaxPooling2D(pool_size=(2, 2), name='max_pooling2d_1'),
    
    Conv2D(64, (3, 3), activation='relu', name='conv2d_2'),
    BatchNormalization(name='batch_norm_2'),
    MaxPooling2D(pool_size=(2, 2), name='max_pooling2d_2'),
    
    Conv2D(128, (3, 3), activation='relu', name='conv2d_3'),
    BatchNormalization(name='batch_norm_3'),
    MaxPooling2D(pool_size=(2, 2), name='max_pooling2d_3'),
    
    Conv2D(256, (3, 3), activation='relu', name='conv2d_4'),
    BatchNormalization(name='batch_norm_4'),
    MaxPooling2D(pool_size=(2, 2), name='max_pooling2d_4'),
    
    Dropout(0.5, name='dropout_1'),
    GlobalAveragePooling2D(name='global_avg_pooling'),
    
    Dense(128, activation='relu', name='dense_1'),
    Dropout(0.3, name='dropout_2'),
    
    Dense(256, activation='relu', name='dense_2'),
    Dropout(0.3, name='dropout_3'),
    
    Dense(6, activation='softmax', name='output_dense')
])

# Generate model summary
model.summary()

# Plot the new model architecture
plot_model(model, to_file="cnn_model_plot.png", show_shapes=True, show_layer_names=True)
```
